```python
from phi.agent import Agent, RunResponse
from phi.model.groq import Groq
from phi.tools.duckduckgo import DuckDuckGo
import time

# Create agent with Groq's Llama-3.1-8b-instant
web_agent = Agent(
    name="Web Agent",
    model=Groq(
        id="llama-3.1-8b-instant",  # Using Llama 3.1 8B Instant model from Groq
        api_key="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"  # Replace with your actual Groq API key
    ),
    tools=[DuckDuckGo()],
    instructions=[
        "You are a helpful assistant with access to search tools.",
        "Always use the search tool when asked about current events or topics you might not know about.",
        "Always include sources",
        "Provide 2-3 course recommendations per category with links."
    ],
    show_tool_calls=True,
    markdown=True,
)

# Function to run queries with delay to avoid rate limiting
def run_with_delay(query, delay_seconds=10):
    print(f"Searching for: {query}")
    result = web_agent.run(query)
    print(f"Completed search for: {query}")
    print(f"Waiting {delay_seconds} seconds before next query...")
    time.sleep(delay_seconds)  # Add delay between queries
    return result

# Run separate queries for different skill levels with delays
foundational_results = run_with_delay("Find recommended foundational Python courses for beginners with their links and brief descriptions")
print("\n=== FOUNDATIONAL PYTHON COURSES ===\n")
print(foundational_results.content)

# Longer delay between major queries
time.sleep(20)

intermediate_results = run_with_delay("Find recommended intermediate Python courses with their links and brief descriptions")
print("\n=== INTERMEDIATE PYTHON COURSES ===\n")
print(intermediate_results.content)

# Longer delay between major queries
time.sleep(20)

advanced_results = run_with_delay("Find recommended advanced Python courses focusing on topics like data science, machine learning or backend development with their links")
print("\n=== ADVANCED PYTHON COURSES ===\n")
print(advanced_results.content)
```

```python
from phi.agent import Agent
from phi.model.groq import Groq
import time
import random
import json

# Create agent with Groq's Llama-3.1-8b-instant but WITHOUT DuckDuckGo
groq_agent = Agent(
    name="Course Recommender",
    model=Groq(
        id="llama-3.1-8b-instant",  # Using Llama 3.1 8B Instant model from Groq
        api_key="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"  # Your Groq API key
    ),
    instructions=[
        "You are a helpful assistant specializing in course recommendations.",
        "For each course request, provide exactly the number requested for the given topic.",
        "For each course, provide:",
        "1. The exact course name",
        "2. Its direct link (use well-known legitimate course platforms like Coursera, Udemy, etc.)",
        "3. A brief 1-2 sentence description of what the course covers",
        "ONLY include courses that are specifically designed for the requested skill level.",
        "Format your response as a numbered list with clear separation between courses.",
        "Avoid unnecessary commentary before or after the list."
    ],
    show_tool_calls=False,  # Hide tool calls for cleaner output
    markdown=True,
)

# Function to run queries with randomized delays to avoid rate limiting patterns
def run_with_smart_delay(query, min_delay=5, max_delay=10):
    print(f"Generating recommendations for: {query}")

    try:
        result = groq_agent.run(query)
        print(f"✓ Recommendations generated successfully")
        return result.content
    except Exception as e:
        print(f"✗ Generation failed: {str(e)}")
        return f"Error: Could not generate recommendations. Error: {str(e)}"
    finally:
        delay = random.randint(min_delay, max_delay)
        print(f"Waiting {delay} seconds before next query...")
        time.sleep(delay)

# Parse results to extract structured course information
def parse_course_results(text):
    courses = []
    lines = text.strip().split('\n')
    current_course = {}

    for line in lines:
        line = line.strip()
        if not line:
            continue

        # Check for new course entry (starts with number)
        if line.startswith('1.') or line.startswith('2.') or line.startswith('3.'):
            # Save previous course if exists
            if current_course and 'name' in current_course:
                courses.append(current_course)
                current_course = {}

            # Extract course name
            parts = line.split('.', 1)
            if len(parts) > 1:
                current_course['name'] = parts[1].strip()

        # Look for link (contains http)
        elif 'http' in line.lower() and 'name' in current_course and 'link' not in current_course:
            current_course['link'] = line.strip()

        # If we have name and link but no description yet, this must be the description
        elif 'name' in current_course and 'link' in current_course and 'description' not in current_course:
            current_course['description'] = line.strip()

    # Add the last course if it exists
    if current_course and 'name' in current_course:
        courses.append(current_course)

    return courses

# More specific and targeted queries for each category
print("\n=== FOUNDATIONAL COURSES ===\n")
beginner_courses = run_with_smart_delay(
    """Provide exactly 5 courses:
    1. One Python course for complete beginners
    2. One JavaScript course for complete beginners
    3. One Java course for complete beginners
    4. One Machine Learning course for beginners
    5. One Deep Learning course for beginners

    For each, provide the course name, a legitimate link, and a brief description."""
)
print(beginner_courses)

print("\n=== INTERMEDIATE COURSES ===\n")
intermediate_courses = run_with_smart_delay(
    """Provide exactly 3 intermediate-level courses:
    1. One intermediate Python course focusing on web development
    2. One intermediate Python course focusing on data analysis
    3. One intermediate Python course focusing on automation

    For each, provide the course name, a legitimate link, and a brief description."""
)
print(intermediate_courses)

print("\n=== ADVANCED COURSES ===\n")
advanced_courses = run_with_smart_delay(
    """Provide exactly 3 advanced-level Python courses:
    1. One advanced Python course focusing on data science
    2. One advanced Python course focusing on machine learning
    3. One advanced Python course focusing on backend development

    For each, provide the course name, a legitimate link, and a brief description."""
)
print(advanced_courses)

# Convert all results to structured JSON
try:
    beginner_parsed = parse_course_results(beginner_courses)
    intermediate_parsed = parse_course_results(intermediate_courses)
    advanced_parsed = parse_course_results(advanced_courses)

    all_courses = {
        "beginner": beginner_parsed,
        "intermediate": intermediate_parsed,
        "advanced": advanced_parsed
    }

    # Save results to JSON file
    with open('course_recommendations.json', 'w') as f:
        json.dump(all_courses, f, indent=2)

    print("\n✓ All course recommendations saved to course_recommendations.json")
    print("\nSample of JSON structure:")
    print(json.dumps(all_courses, indent=2)[:500] + "...")

except Exception as e:
    print(f"\n✗ Error processing results: {str(e)}")
```
